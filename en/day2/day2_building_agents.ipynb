### 🔧 Translated English text Async + Runner?

- **`async/await`** — ADK Translated English text asynchronous Translated English text Agent Translated English text LLM Translated English text (I/O bound)
- **`InMemoryRunner`** — Translated English text Agent Translated English text event loop Translated English text
- **`InMemorySessionService`** — Translated English text RAM

```
Runner → Translated English text Session → Translated English text Agent → Agent Translated English text
  → Translated English text Tool (Translated English text) → Translated English text → Translated English text Event Translated English text
```

> 💡 Translated English text Colab Translated English text `await` Translated English text cell (Translated English text `asyncio.run()`)

````python
# ─── Translated English text Agent Translated English text Colab ───
import asyncio

async def chat_with_agent(runner, message, user_id='user_1', session_id='default_session'):
    """Translated English text Agent Translated English text"""
    content = types.Content(
        role='user',
        parts=[types.Part.from_text(text=message)]
    )
    
    final_response = ''
    tool_calls = []
    
    async for event in runner.run_async(
        user_id=user_id,
        session_id=session_id,
        new_message=content
    ):
        if event.content and event.content.parts:
            for part in event.content.parts:
                if part.text:
                    final_response = part.text
                if part.function_call:
                    tool_calls.append({
                        'name': part.function_call.name,
                        'args': dict(part.function_call.args) if part.function_call.args else {}
                    })
    
    if tool_calls:
        print(f'  🔧 Tools Translated English text: {[t["name"] for t in tool_calls]}')
    
    return final_response


async def demo_chat(agent, messages):
    """Demo Translated English text Agent (Translated English text session Translated English text)"""
    runner = InMemoryRunner(agent=agent, app_name=agent.name)
    session_id = f'session_{id(agent)}'
    
    # Translated English text session Translated English text runner
    await runner.session_service.create_session(
        app_name=agent.name,
        user_id='user_1',
        session_id=session_id
    )
    
    for msg in messages:
        print(f'\n👤 Translated English text: {msg}')
        response = await chat_with_agent(runner, msg, session_id=session_id)
        print(f'🤖 Agent: {response}')


print('✅ Translated English text chat Translated English text')
````

### Tool Translated English text?

**Tool = Python function Translated English text Agent Translated English text**


### 📌 Translated English text Tool

| Translated English text | Translated English text? |
|-----|-------|
| Translated English text **docstring** | LLM Translated English text docstring Translated English text tool |
| Translated English text **type hints** | LLM Translated English text parameter Translated English text |

````python
%%time
# ─── Tool 1: Translated English text BMI ───
def calculate_bmi(weight_kg: float, height_cm: float) -> dict:
    """Translated English text (BMI) Translated English text

    Args:
        weight_kg: Translated English text
        height_cm: Translated English text

    Returns:
        dict: Translated English text BMI, Translated English text, Translated English text
    """
    height_m = height_cm / 100
    bmi = weight_kg / (height_m ** 2)

    if bmi < 18.5:
        level = 'Translated English text'
        advice = 'Translated English text'
    elif bmi < 25:
        level = 'Translated English text'
        advice = 'Translated English text'
    elif bmi < 30:
        level = 'Translated English text'
        advice = 'Translated English text'
    else:
        level = 'Translated English text'
        advice = 'Translated English text'

    return {'bmi': round(bmi, 1), 'level': level, 'advice': advice}


# ─── Tool 2: Translated English text ───
def convert_temperature(value: float, from_unit: str) -> dict:
    """Translated English text Celsius Translated English text Fahrenheit

    Args:
        value: Translated English text
        from_unit: Translated English text ('celsius' Translated English text 'fahrenheit')

    Returns:
        dict: Translated English text
    """
    if from_unit.lower() == 'celsius':
        result = (value * 9/5) + 32
        return {'result': round(result, 1), 'from': f'{value}°C', 'to': f'{result:.1f}°F'}
    elif from_unit.lower() == 'fahrenheit':
        result = (value - 32) * 5/9
        return {'result': round(result, 1), 'from': f'{value}°F', 'to': f'{result:.1f}°C'}
    else:
        return {'error': f'Translated English text {from_unit}'}


# ─── Translated English text Agent Translated English text Tools ───
health_agent = Agent(
    name='health_agent',
    model='gemini-2.5-flash',
    description='Agent Translated English text BMI Translated English text',
    instruction="""Translated English text Agent Translated English textThai language
    - Translated English text calculate_bmi Translated English text BMI Translated English text
    - Translated English text convert_temperature Translated English text
    - Translated English text tools Translated English text
    """,
    tools=[calculate_bmi, convert_temperature]
)

print('✅ Translated English text Health Agent Translated English text 2 tools')
print(f'   🔧 Tools: {[t.__name__ for t in health_agent.tools]}')
````

````python
%%time
# ─── Translated English text Qdrant VectorDB (Translated English text Day 1) ───
import hashlib, random
from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer
from qdrant_client import QdrantClient, models

# Translated English text Embedding Model
embed_model = SentenceTransformer('intfloat/multilingual-e5-large')
print('✅ Translated English text embedding model')

# Translated English text Day 1
documents = {
    'healthcare': 'Translated English text AI Translated English text X-ray Translated English text 30 Translated English text 5 Translated English text Deep Learning Translated English text 95% Translated English text',
    'banking': 'Translated English text KADE AI Assistant Translated English text RAG Translated English text 5 Translated English text 30 Translated English textThai languageTranslated English text',
    'education': 'Translated English text RAG Translated English text-Translated English text PDF Translated English text 500 Translated English text vector embeddings Translated English text multilingual model Translated English textThai language studentsTranslated English text',
    'agriculture': 'Translated English text AI Translated English text Computer Vision Translated English text 92% Translated English text 40%',
    'kmitl': 'Translated English text (KMITL) Translated English text AI Translated English text Smart Campus Translated English text IoT sensor Translated English text Machine Learning Translated English text 25% Translated English text NLP Thai languageTranslated English textstudents Translated English text RAG Translated English text AI Tutor Translated English text',
    'nlp': 'Translated English text NLP Translated English textThai languageTranslated English textThai languageTranslated English text Word Segmentation Translated English text PyThaiNLP Translated English text tokenization',
    'rag_architecture': 'Translated English text RAG Translated English text 3 Translated English text Retriever Translated English text VectorDB, Reader Translated English text, Translated English text Generator Translated English text RAG Translated English text hallucination Translated English text LLM Translated English text',
    'embedding': 'Text Embedding Translated English text vector Translated English text multilingual-e5-large Translated English text Microsoft Translated English text 100 Translated English textThai language Translated English text vector Translated English text 1024 Translated English text vector Translated English text embedding',
    'vector_db': 'Qdrant Translated English text Vector Database Translated English text ANN (Approximate Nearest Neighbor) Translated English text vector Translated English text Cosine Similarity, Dot Product, Translated English text Euclidean Distance',
}

# Chunk + Embed + Index
splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=30)

all_chunks = []
all_sources = []
for source, text in documents.items():
    chunks = splitter.split_text(text)
    for chunk in chunks:
        all_chunks.append(chunk)
        all_sources.append(source)

print(f'📊 {len(all_chunks)} chunks Translated English text {len(documents)} Translated English text')

# Embed
passages = [f'passage: {c}' for c in all_chunks]
embeddings = embed_model.encode(passages, show_progress_bar=True)
VECTOR_SIZE = embeddings.shape[1]
print(f'✅ Embedding: {embeddings.shape}')

# Create Qdrant collection
qdrant = QdrantClient(':memory:')
qdrant.create_collection(
    collection_name='thai_ai_kb',
    vectors_config=models.VectorParams(size=VECTOR_SIZE, distance=models.Distance.COSINE)
)

# Upsert
points = [
    models.PointStruct(
        id=i,
        vector=embeddings[i].tolist(),
        payload={'text': all_chunks[i], 'source': all_sources[i]}
    )
    for i in range(len(all_chunks))
]
qdrant.upsert(collection_name='thai_ai_kb', points=points)
print(f'✅ Upsert {len(points)} points Translated English text Qdrant')
````

````python
%%time
# ─── Translated English text RAG Tool ───
def search_thai_ai_knowledge(query: str) -> dict:
    """Translated English text AI Translated English text

    Translated English text tool Translated English text:
    - Translated English text AI Translated English text (Translated English text)
    - Translated English text RAG, Embedding, Vector Database
    - NLP Thai language

    Args:
        query: Translated English textThai languageTranslated English text

    Returns:
        dict: Translated English text
    """
    query_vector = embed_model.encode(f'query: {query}').tolist()
    results = qdrant.query_points(
        collection_name='thai_ai_kb',
        query=query_vector,
        limit=3
    ).points

    if not results:
        return {'status': 'no_results', 'message': 'Translated English text'}

    contexts = []
    for r in results:
        contexts.append({
            'text': r.payload['text'],
            'source': r.payload['source'],
            'score': round(r.score, 4)
        })

    return {
        'status': 'success',
        'num_results': len(contexts),
        'results': contexts
    }

print('✅ RAG Tool Translated English text')

# Translated English text tool Translated English text
test = search_thai_ai_knowledge('AI Translated English text')
print(f'\n🧪 Translated English text: {test["num_results"]} Translated English text')
for r in test['results']:
    print(f'  [{r["source"]}] score={r["score"]} | {r["text"][:60]}...')
````

````python
%%time
# ─── Translated English text RAG Agent ───
rag_agent = Agent(
    name='thai_ai_expert',
    model='gemini-2.5-flash',
    description='Translated English text AI Translated English text',
    instruction="""Translated English text AI Translated English text

    Translated English text:
    1. Translated English text AI Translated English text, case study, Translated English text RAG
       → Translated English text search_thai_ai_knowledge Translated English text
    2. Translated English text
    3. Translated English text "Translated English text"
    4. Translated English text → Translated English text
    5. Translated English textThai language Translated English text bullet points
    """,
    tools=[search_thai_ai_knowledge]
)

print('✅ Translated English text RAG Agent Translated English text!')
````

### 3️⃣ LoopAgent — Translated English text

| Step | Action | Result |
|:----:|--------|--------|
| 1 | Translated English text | draft |
| 2 | Translated English text | Translated English text? → Translated English text Step 1 |
| 3 | Translated English text | ✅ Translated English text |

**Translated English text self-improvement**: Agent Translated English text → Agent Translated English text → Translated English text

> ⚠️ Translated English text `max_iterations` Translated English text!

````python
%%time
from google.adk.agents import LoopAgent

# Agent 1: Translated English text
writer_agent = Agent(
    name='writer_agent',
    model='gemini-2.5-flash',
    description='Translated English textcontent',
    instruction="""Translated English textcontentTranslated English text
    Translated English text feedback Translated English text""",
    tools=[search_thai_ai_knowledge]
)

# Agent 2: Translated English text (critic)
critic_agent = Agent(
    name='critic_agent',
    model='gemini-2.5-flash',
    description='Translated English text feedback',
    instruction="""Translated English text:
    - Translated English text?
    - Translated English text?
    - Translated English text?
    
    Translated English text → Translated English text
    Translated English text → Translated English text 'APPROVED: [Translated English text]'
    """
)

# LoopAgent: Translated English text → Translated English text → Translated English text → Translated English text → ... (Translated English text 3 Translated English text)
loop_refiner = LoopAgent(
    name='refine_loop',
    description='Translated English text',
    sub_agents=[writer_agent, critic_agent],
    max_iterations=3  # Translated English text!
)

print('✅ Translated English text LoopAgent: Translated English text ↔ Translated English text (Translated English text 3 Translated English text)')
print(f'   Max iterations: {loop_refiner.max_iterations}')
````

````python
runner = InMemoryRunner(agent=loop_refiner, app_name=loop_refiner.name)
await runner.session_service.create_session(app_name=loop_refiner.name, user_id='user_1', session_id='test_session')
# ─── Translated English text Loop Agent ───
print('═' * 60)
print('🧪 LoopAgent: Translated English text → Translated English text → Translated English text → Translated English text ...')
print('═' * 60)

response = await chat_with_agent(runner, 'Translated English text RAG Translated English text', session_id='test_session')
print(f'\n📋 Translated English text (Translated English text):')
print(response)
````
