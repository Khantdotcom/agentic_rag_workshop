## 📦 Setup (Run Translated English text)

```python
%%time
!pip install -q google-genai requests nbformat langchain-text-splitters

import hashlib, os, json, random, re, csv, requests
from datetime import datetime
import nbformat
from google import genai

# Translated English text Gemini API
try:
    from google.colab import userdata
    GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
    print('✅ Translated English text API Key Translated English text Colab Secrets')
except Exception:
    GOOGLE_API_KEY = input('🔑 Translated English text Gemini API Key: ')

client = genai.Client(api_key=GOOGLE_API_KEY)
MODEL = 'gemini-2.5-pro'

# Translated English textpoints (append mode)
SCORES_CSV = 'day1_scores.csv'
SCORES_JSON = 'day1_scores.json'

# Translated English text CSV header Translated English text
if not os.path.exists(SCORES_CSV):
    with open(SCORES_CSV, 'w', encoding='utf-8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['Timestamp', 'Student Name', 'Student ID', 'Phone', 'LINE ID',
                         'Step1', 'Step2', 'Step3', 'Step4', 'Total',
                         'AI Suspected', 'Feedback', 'GitHub URL'])
    print(f'📄 Translated English text {SCORES_CSV} Translated English text')
else:
    # Translated English text
    with open(SCORES_CSV, 'r') as f:
        existing = sum(1 for _ in f) - 1
    print(f'📄 {SCORES_CSV} Translated English text {existing} Translated English text')

# Translated English text JSON Translated English text
if not os.path.exists(SCORES_JSON):
    with open(SCORES_JSON, 'w') as f:
        json.dump([], f)

print(f'✅ Setup Translated English text | Model: {MODEL}')
```

## 🔧 Translated English text (Run Translated English text)

```python
def generate_expected_answers(student_id):
    """Translated English textstudents"""
    from langchain_text_splitters import RecursiveCharacterTextSplitter
    
    random.seed(int(hashlib.md5(student_id.encode()).hexdigest()[:8], 16))
    
    all_paragraphs = [
        'Translated English text Machine Learning Translated English text',
        'Deep Learning Translated English text Machine Learning Translated English text Neural Network Translated English text',
        'Natural Language Processing Translated English text NLP Translated English text',
        'Retrieval Augmented Generation Translated English text RAG Translated English text LLM Translated English text',
        'Vector Database Translated English text Embedding Vector Translated English text',
        'Text Embedding Translated English text Vector Translated English text',
        'Transformer Translated English text Neural Network Translated English text Attention Translated English text GPT BERT Translated English text Gemini',
        'Prompt Engineering Translated English text Prompt Translated English text LLM Translated English text Prompt Translated English text',
        'Fine-tuning Translated English text',
        'Tokenization Translated English text Token Translated English textThai languageTranslated English text',
        'Chunking Translated English text Embedding Translated English text Fixed-size Recursive Translated English text Semantic Chunking',
        'Cosine Similarity Translated English text Vector Translated English text Vector Translated English text 1 Translated English text 0 Translated English text NLP Translated English text Information Retrieval',
    ]
    
    random.shuffle(all_paragraphs)
    selected = all_paragraphs[:8]
    duplicate_idx = random.randint(0, 5)
    selected.append(selected[duplicate_idx])
    
    hashes = {}
    dup_files = None
    dup_hash = None
    for i, para in enumerate(selected):
        h = hashlib.md5(para.encode()).hexdigest()
        fname = f'doc_{i+1}.txt'
        if h in hashes:
            dup_files = (hashes[h], fname)
            dup_hash = h
        else:
            hashes[h] = fname
    
    unique_texts = []
    seen = set()
    for para in selected:
        h = hashlib.md5(para.encode()).hexdigest()
        if h not in seen:
            unique_texts.append(para)
            seen.add(h)
    
    all_text = '\n\n'.join(unique_texts)
    splitter = RecursiveCharacterTextSplitter(chunk_size=150, chunk_overlap=30)
    chunks = splitter.split_text(all_text)
    
    return {
        'student_id': student_id,
        'dup_files': dup_files,
        'dup_hash': dup_hash,
        'files_after_dedup': len(unique_texts),
        'num_chunks': len(chunks),
        'chunk_1_content': chunks[0] if chunks else '',
        'min_chunk_len': min(len(c) for c in chunks) if chunks else 0,
        'verification_code': hashlib.sha256(f'{student_id}_day1_hw'.encode()).hexdigest()[:12],
    }


def fetch_notebook(github_url):
    """Translated English text notebook + Translated English textstudentsTranslated English text GitHub"""
    raw_url = github_url.replace('github.com', 'raw.githubusercontent.com').replace('/blob/', '/')
    print(f'📥 Translated English text: {raw_url}')
    resp = requests.get(raw_url)
    resp.raise_for_status()
    nb = nbformat.reads(resp.text, as_version=4)
    
    info = {'student_name': '', 'student_id': '', 'phone': '', 'line_id': ''}
    full_content = ''
    
    for i, cell in enumerate(nb.cells):
        if cell.cell_type == 'code':
            for key, var in [('student_id','STUDENT_ID'), ('student_name','STUDENT_NAME'),
                             ('phone','PHONE'), ('line_id','LINE_ID')]:
                m = re.search(rf"{var}\s*=\s*['\"]([^'\"]+)['\"]", cell.source)
                if m:
                    info[key] = m.group(1)
        
        ct = cell.cell_type.upper()
        full_content += f'\n\n=== CELL {i} ({ct}) ===\n{cell.source}'
        if hasattr(cell, 'outputs'):
            for out in cell.outputs:
                if hasattr(out, 'text'):
                    full_content += f'\n--- OUTPUT ---\n{out.text}'
                elif hasattr(out, 'data'):
                    for mime, data in out.data.items():
                        if 'text' in mime:
                            txt = data if isinstance(data, str) else '\n'.join(data)
                            full_content += f'\n--- OUTPUT ---\n{txt}'
    
    return info, full_content


def grade_with_gemini(student_id, notebook_content, expected):
    """Translated English text Gemini 2.5 Pro Translated English textpoints"""
    prompt = f'''Translated English texthomeworkTranslated English text AI/RAG Translated English textstudentsundergraduate

# Translated English text
- Translated English text: {expected["dup_files"]}, MD5: {expected["dup_hash"]}, Translated English text {expected["files_after_dedup"]} Translated English text
- Translated English text chunks: {expected["num_chunks"]}, chunk Translated English text: {expected["min_chunk_len"]} Translated English text
- chunk Translated English text (80 Translated English text): "{expected["chunk_1_content"][:80]}..."
- Verification: {expected["verification_code"]}

# Notebook
{notebook_content[:15000]}

# Translated English text
## Step 1: MD5 (2 points) — 1 Translated English text, 0.5 hash Translated English text, 0.5 Translated English text
## Step 2: Chunking (2 points) — 1 Translated English text chunks Translated English text, 0.5 content chunk 1 Translated English text, 0.5 chunk Translated English text
## Step 3: Embedding (3 points) — 1 Translated English text multilingual-e5-large, 1 Translated English text cosine similarity+score, 1 Translated English text
## Step 4: Qdrant (3 points) — 1 Translated English text collection+upsert, 1 Top-3 scores, 1 Translated English text

Translated English text JSON:
{{"step1_score":0.0,"step1_feedback":"...","step2_score":0.0,"step2_feedback":"...","step3_score":0.0,"step3_feedback":"...","step4_score":0.0,"step4_feedback":"...","total_score":0.0,"overall_feedback":"...","is_ai_suspected":false,"ai_reason":"..."}}

Translated English text feedback Translated English textThai language Translated English text
Translated English text code/output Translated English text 0 points
Translated English text AI Translated English text is_ai_suspected=true
'''
    resp = client.models.generate_content(
        model=MODEL, contents=prompt,
        config=genai.types.GenerateContentConfig(temperature=0.1, response_mime_type='application/json')
    )
    return json.loads(resp.text)


def append_result(info, grade, github_url):
    """Append Translated English textpointsTranslated English text CSV + JSON"""
    ts = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    # Append CSV
    with open(SCORES_CSV, 'a', encoding='utf-8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([
            ts, info['student_name'], info['student_id'], info['phone'], info['line_id'],
            grade['step1_score'], grade['step2_score'], grade['step3_score'], grade['step4_score'],
            grade['total_score'], grade.get('is_ai_suspected', False),
            grade['overall_feedback'], github_url
        ])
    
    # Append JSON
    with open(SCORES_JSON, 'r') as f:
        all_results = json.load(f)
    grade.update({'timestamp': ts, 'github_url': github_url, **info})
    all_results.append(grade)
    with open(SCORES_JSON, 'w', encoding='utf-8') as f:
        json.dump(all_results, f, ensure_ascii=False, indent=2)
    
    # Translated English text
    print(f'💾 Translated English text (Translated English text {len(all_results)} Translated English text)')


print('✅ Translated English text')
```

```python
assert GITHUB_URL, '❌ Translated English text GitHub URL!'

# 1. Translated English text notebook
info, content = fetch_notebook(GITHUB_URL)
print(f'👤 {info["student_name"]} ({info["student_id"]})')
print(f'📱 {info["phone"]} | 💬 {info["line_id"]}')

# 🔍 Translated English text
already_graded = False
if os.path.exists(SCORES_JSON):
    with open(SCORES_JSON, 'r') as f:
        prev = json.load(f)
    for p in prev:
        if p.get('student_id') == info['student_id']:
            already_graded = True
            print(f'⚠️  Translated English text {info["student_id"]} Translated English text (points: {p.get("total_score", "?")})')
            break

if already_graded:
    overwrite = input('🔄 Translated English text overwrite pointsTranslated English text? (y/n): ').strip().lower()
    if overwrite != 'y':
        print('⏭️  Translated English textstudentsTranslated English text')
        raise SystemExit()
    # Translated English text
    prev = [p for p in prev if p.get('student_id') != info['student_id']]
    with open(SCORES_JSON, 'w', encoding='utf-8') as f:
        json.dump(prev, f, ensure_ascii=False, indent=2)
    # Translated English text CSV Translated English text
    import csv
    rows = []
    with open(SCORES_CSV, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        rows = list(reader)
    with open(SCORES_CSV, 'w', encoding='utf-8', newline='') as f:
        writer = csv.writer(f)
        for row in rows:
            if len(row) > 2 and row[2] == info['student_id']:
                continue
            writer.writerow(row)
    print(f'🗑️  Translated English textpointsTranslated English text {info["student_id"]} Translated English text')

# 2. Translated English text
expected = generate_expected_answers(info['student_id'])
print(f'🔑 Translated English text: dup={expected["dup_files"]}, chunks={expected["num_chunks"]}')

# 3. Translated English text Gemini
print(f'🤖 Translated English text {MODEL}...')
grade = grade_with_gemini(info['student_id'], content, expected)

# 4. Translated English text
print(f'\n{"="*60}')
print(f'📊 Translated English text: {info["student_name"]} ({info["student_id"]})')
print(f'{"="*60}')
print(f'  Translated English text 1 (MD5):       {grade["step1_score"]}/2  — {grade["step1_feedback"]}')
print(f'  Translated English text 2 (Chunking):  {grade["step2_score"]}/2  — {grade["step2_feedback"]}')
print(f'  Translated English text 3 (Embedding): {grade["step3_score"]}/3  — {grade["step3_feedback"]}')
print(f'  Translated English text 4 (Qdrant):    {grade["step4_score"]}/3  — {grade["step4_feedback"]}')
print(f'  {"─"*56}')
print(f'  🏆 pointsTranslated English text: {grade["total_score"]}/10')
print(f'  💬 {grade["overall_feedback"]}')
if grade.get('is_ai_suspected'):
    print(f'  ⚠️  Translated English text AI: {grade["ai_reason"]}')

# 5. Append Translated English text file
append_result(info, grade, GITHUB_URL)
```
